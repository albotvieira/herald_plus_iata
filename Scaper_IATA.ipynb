```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
```

## Iterate over search pages

```python
import re

def get_total_n_pages(url):

    r = requests.get(url)

    # Convert to beautiful soup object 
    webpage = bs(r.content)

    ulist = webpage.find("div", attrs={"class": "pagination"})
    # Set big one string to list of elements
    links = ulist.find_all("a")
    links

    actual_links = [link['data-search-nav-url-with-form'] for link in links]
    actual_links

    regexp = re.compile("page=(.*)$")

    ref_val = 0
    for id in range(0,len(actual_links)-1):

        curr_val = int( regexp.search( actual_links[id]).group(1) )
        if curr_val > ref_val:
            ref_val = curr_val

    return ref_val
```

## Main Data fetcher

```python
# Load the webpage content 

def parse_html_content(url):
    r = requests.get(url)

    # Convert to beautiful soup object 
    soup = bs(r.content)

    # Print our 
    #print(soup.prettify() )
    
    '''
    for span in soup.findAll("table", class_ = "datatable"):#, id = "headline_avherald" ):
        #print( span.text )
    
    for span in soup.findAll("table", class_ = "datatable"):#, id = "headline_avherald" ):
        #print( span.text )
    '''
    table = soup.find( "table", {"class":"datatable"} )

    rows=list()
    for row in table.findAll("tr"):    

        row_parser = row.text.split('\n')
        row_output = [i for i in row_parser if i]
        #print( row_output  )

        rows.append( row_output)
    
    return rows    
```

```python
# Load parsed data into panda dataframe (table)
def load_data_into_df(list):
    df = pd.DataFrame( rows[1:], columns = rows[0])
    #df.head(5)
    
    return df
```

## Main function

```python
url = "https://www.iata.org/en/about/members/airline-list/"

n_pages = get_total_n_pages(url)
#n_pages = 1
print( "Num.pages = ", n_pages)

my_df = pd.DataFrame()

for i in range(1, n_pages+1 ):
    rolling_url = "https://www.iata.org/en/about/members/airline-list/?page=" + str(i) + "&search=&ordering=Alphabetical"
    print(rolling_url)
    
    my_df = my_df.append( load_data_into_df(  parse_html_content(rolling_url) ) ) 
```
