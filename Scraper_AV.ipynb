```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
```

#### Load our first page

```python

def parse_html_content(url):
    # Load the webpage content 
    r = requests.get(url)

    # Convert to beautiful soup object 
    soup = bs(r.content)

    # Print our 
    #print(soup.prettify() )

    #for span in soup.findAll("span"):
        #print( span )
        
    my_raw, my_type = [], []
    
    for span in soup.findAll("span", class_ = "headline_avherald"):
        #print( span.text )
        my_raw.append(span.text)

    for img in soup.findAll("img", alt = ['Crash', 'Accident', 'Incident']): #**still to find other categories
        #print( img['title'] )
        my_type.append(img['title'])  

    return [my_raw, my_type]
```

```python

def load_data_into_df( lists):
    my_raw = lists[0]
    my_type = lists[1]
    df = pd.DataFrame( list(zip(my_type, my_raw )), columns = ['Type_of_incident','Raw'])
    #df.head()

    return df
```

```python
url = "http://avherald.com/"

load_data_into_df(  parse_html_content(url)).head(10)

```
